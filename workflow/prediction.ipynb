# Define collisional system parameters

To use this package you need to provide the following parameters:
    
    - projectile: Chemical formula for the projectile
    - projectile_mass: Mass in amu for the projectile
    - target: Chemical formula for the target
    - target_mass: Mass in amu for the target

The package produces a sample file named ``system_prediction.csv``, where ``system`` is the name of the collisional system given.

```python
SPNN.run_SPNN(projectile='S', projectile_mass=1, target='He', target_mass=2)
```

## Optional: incident energy grid in a logarithmic scale (10 ** x) of MeV/amu

    - minlogE : exponent value of minimum grid point
    - maxlogE : exponent value of maximum grid point
    - npoints : number of points in grid
    - fdir: folder path to output file

```python
SPNN.run_SPNN(projectile='He', projectile_mass=2, target='Au', target_mass=196.966569, minlogE=-2, maxlogE=1, npoints=50)
```

The minimal input to use the package from your terminal is given by the following syntaxis:

$ python -m SPNN Li 6.941 Co 63.546
