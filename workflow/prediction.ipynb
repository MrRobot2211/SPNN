```python
import SPNN
```

# Define collisional system parameters

To use this package you need to provide the following parameters:
    
- ``projectile``: Chemical formula for the projectile
- ``projectile_mass``: Mass in amu for the projectile
- ``target``: Chemical formula for the target
- ``target_mass``: Mass in amu for the target

The package produces a sample file named ``system_prediction.dat``, where ``system`` is the name of the collisional system given.

```python
SPNN.run_SPNN(projectile='H', projectile_mass=1.0, target='He', target_mass=4.002602)
```

## Optional: incident energy grid in MeV/amu units

- ``emin``: Minimum energy value (default: 0.001)
- ``emax``: Maximum energy value (default: 10)
- ``npoints``: Number of grid points (default: 1000)
- ``plot``: Plot prediction (default: True)
- ``outdir``: Path to output folder (default: "./")

```python
SPNN.run_SPNN(projectile='He', projectile_mass=4.002602, target='Au', target_mass=196.966569, emin=0.01, emax=1, npoints=50)
```

# Run SPNN in terminal

The minimal input to use the SPNN package from your terminal is given by the following syntaxis:

``$ python -m SPNN Li 6.941 Co 63.546 ``

More info about the package can be obtained with the -h, --help flag

``$ python -m SPNN -h``
